### Queries using PySpark DSL

DSL = Domain Specific Language - API similar to natural or other language, implemented as library in another language.

List all the countries with the population > 38 million

```python
countries.filter((col("Population") > 38000000)).orderBy("Population").toPandas()
```

Select all the countries from Europe

```python
countries.select("Country", "Population").where(col("Region").like("%EUROPE%")).show()
```

Conditions in `where` clause can contain logical expressions.

```python
countries.select("Country", "Population")\
.where((col("Region").like("%EUROPE%")) & (col("Population")> 10000000)).show()
```

## User defined functions for data manipulation
Our `countries` DataFrame has some problems:
 * missing values
 * some numbers use comma instead of point as floating point separator (e.g. Literacy  = 99,4)
 
We can clean the data using User Defined Functions (UDF)

```python
from pyspark.sql.types import FloatType
from pyspark.sql.functions import udf
```

Define a Python function which coverts numbers with commas to `float`

```python
def to_float(s):
    return float(s.replace(',','.'))
```

Test that it works:

```python
to_float('0,99')
```

Now define a Spark UDF:

```python
float_udf = udf(to_float , FloatType())
```

Test it on a Data Frame

```python
countries.withColumn("Literacy", float_udf("Literacy (%)"))
```

OK, we can see that the `Literacy` is now `float`

```python
countries.show(50)
```

```python
countries.where((col("Literacy") < 50) & (col("GDP ($ per capita)") > 700)).show()
```

Oops, what does it mean???

- some rows have empty values!

Before we can use the table, we need to remove empty rows. Otherwise our UDF will fail.

```python
full_countries = countries.select('Country', 'Population', 'Literacy (%)', 'GDP ($ per capita)').na.drop()
```

We can now apply the new UDF to the Data Frame:

```python
full_countries = full_countries.withColumn("Literacy", float_udf("Literacy (%)"))
```

```python
full_countries.show(50)
```

```python
full_countries.where((col("Literacy") < 50) & (col("GDP ($ per capita)") > 700)).show()
```

```python
full_countries.toPandas().plot(x="Literacy",y="GDP ($ per capita)",kind="scatter",figsize=(10, 6))
```

### 1. 'Plot Infant Mortality' vs 'Literacy' for top 30 richest and poorest countries

```python
country_file_path = 'countries-of-the-world.csv'
countries_df = spark.read.csv(country_file_path, header=True, inferSchema=True)

countries = countries_df
full_countries = countries\
   .select(
      'Country',
      'Literacy (%)',
      'Infant mortality (per 1000 births)',
      'GDP ($ per capita)'
   ).na.drop()

full_countries = full_countries\
   .withColumn("GDP_per_capita", col('GDP ($ per capita)'))\
   .withColumn("literacy", float_udf("Literacy (%)"))\
   .withColumn('infant_mortality', float_udf('Infant mortality (per 1000 births)'))\
   .drop("Literacy (%)", 'Infant mortality (per 1000 births)', 'GDP ($ per capita)')

richest_countries = full_countries.sort('GDP_per_capita', ascending=False).limit(30)
poorest_countries = full_countries.sort('GDP_per_capita', ascending=True).limit(30)

poorest_countries.toPandas().plot(title='Infant Mortality and Literacy in 30 the poorest countries', x="infant_mortality",y="literacy",kind="scatter",figsize=(10, 6))
richest_countries.toPandas().plot(title='Infant Mortality and Literacy in 30 the richest countries', x="infant_mortality",y="literacy",kind="scatter",figsize=(10, 6))
```

### 2. airports

```python
headers = ['Airport ID','Name','City','Country','IATA','ICAO','Latitude','Longitude','Altitude','Timezone','DST','Tz database time zone','Type','Source']

airports_df = spark.read.option( "delimiter", "," ).csv('airports.csv')
airports_df = airports_df.toDF(*headers)
airports_df = airports_df\
   .na.drop()\
   .withColumn('Latitude', float_udf('Latitude'))\
   .withColumn('Longitude', float_udf('Longitude'))\
```

```python
from shapely.geometry import Point
import geopandas as gpd
airports = airports_df.toPandas()
airports['Coordinates'] = list(zip(airports.Longitude, airports.Latitude))

airports['Coordinates'] = airports['Coordinates'].apply(Point)
gdf = gpd.GeoDataFrame(airports, geometry='Coordinates')

world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
ax = world.plot(color='white', edgecolor='black')

gdf.plot(ax=ax, color='red', markersize=3)
```

### 3. For each country find the nearest to the South Pole airport

```python
from geopy.distance import geodesic
from pyspark.sql.window import Window
import pyspark.sql.functions as F

THE_SOUTH_POLE = (0, -90)

@F.udf(returnType=FloatType())
def geodesic_udf(a):
    return geodesic(a, THE_SOUTH_POLE).m

airports = airports_df

airports = airports\
   .withColumn('distance_to_south_pole', geodesic_udf(F.array("Latitude", "Longitude")))\

# use windowing to get 1 value per group(1 airport per country)
window = Window.partitionBy("Country").orderBy(F.desc('distance_to_south_pole'))
airports = airports\
   .select('*', F.row_number().over(window).alias('rank'))\
   .filter(col('rank') <= 1)

# print on map
airports = airports.toPandas()
airports['Coordinates'] = list(zip(airports.Longitude, airports.Latitude))

airports['Coordinates'] = airports['Coordinates'].apply(Point)
gdf = gpd.GeoDataFrame(airports, geometry='Coordinates')

world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
ax = world.plot(color='white', edgecolor='black')

gdf.plot(ax=ax, color='red', markersize=3)
```

### 4. Number of airports vs country area

```python
from pyspark.sql.functions import count
from pyspark.sql.types import StringType

airports = airports_df
countries = countries_df

normalize_country_name_udf = udf(
   lambda s: s.strip(),
   StringType()
)

countries = countries\
   .withColumnRenamed("Area (sq. mi.)", "Area")\
   .select("Country", "Area")\
   .withColumn("Country", normalize_country_name_udf("Country"))

airports = airports\
   .select("Country")\
   .withColumn("Country", normalize_country_name_udf("Country"))

airports_by_country_area_df = countries\
   .join(airports, on="Country", how="left")\
   .groupBy("Country", "Area")\
   .agg(count("Country"))\
   .drop("Country")\
   .orderBy("Area")

ax = airports_by_country_area_df.toPandas().plot.scatter(
    x="Area",
    y="count(Country)"
)
ax.set_title("Country area vs number of airports")
plt.show()
```

### 5. Airports per continent

```python
airports = airports_df
countries = countries_df

def region_to_continent(s: str) -> str:
    continents_map = {
        "ASIA": "Asia",
        "AFRICA": "Africa",
        "BALTICS": "Europe",
        "EUROPE": "Europe",
        "IND. STATES": "Asia",
        "LATIN AMER": "South America",
        "NEAR EAST": "Africa",
        "NORTHERN AMERICA": "North America",
        "OCEANIA": "Australia/Oceania",
    }

    for key, val in continents_map.items():
        if key in s:
            return val

    return "Other"

region_to_continent_udf = udf(
    lambda s: region_to_continent(s),
    StringType()
)
```

```python
from pyspark.sql.functions import desc

countries = (
    countries
    .select("Country", "Region")
    .withColumn("Country", normalize_country_name_udf("Country"))
    .withColumn("Continent", region_to_continent_udf("Region"))
    .drop("Region")
)

airports = (
    airports
    .select("Country")
    .withColumn("Country", normalize_country_name_udf("Country"))
)

airports_per_continent = (
    countries
    .join(airports, on="Country", how="left")
    .drop("Country")
    .groupBy("Continent")
    .agg(count("Continent"))
    .orderBy(desc("Continent"))
)

ax = airports_per_continent.toPandas().plot.barh(
    x="Continent",
    y="count(Continent)",
    title="Number of airports per continent"
)
plt.show()
```
